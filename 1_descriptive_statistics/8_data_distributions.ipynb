# Data distributions

When I started to study statistics I heard a lot of explanations saying: "This data is normally distributed" or "This data follows the Poisson, binomial, distribution", etc., and even though I got the idea of the bell shape of the normal distribution, etc. I never really understood what a distribution is and how they are used in statistics. Today is another try to organize this knowledge and finally understand this concept with confidence. I hope this can help you as well.

One important chracteristic of a data distribution is its shape. Usually we can use histograms to plot the data and visualize how the data is distributed. Histograms can give us a good clue of the shape of the distribution. 

Statistical inference is based on assumptions about distributions, it's important to undertstand the distribtuion to appropriately apply different inferential statistical methods.

There are many distributions, let's visualize some of them.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
sns.set()
```

## Gaussian aka Normal distribution
- It has a bell shape
- Mean and median are both equal and located at the center of the distribution when it's perfect. If not they will be close to each other around the center
- Approximatelly 68% of the data falls within 1 standard deviation from the mean
- Approximately 95% of the data falls within 2 standard deviations from the mean
- Approximately 99.7% of the data falls within 3 standard deviations from the mean

```python
np.random.seed(42)
mean = 5
variance = 2
x = np.random.normal(mean, variance, 1000)
fig, ax = plt.subplots(figsize=(20,10))
fig.suptitle('Gaussian on Normal distribution', fontsize=32)
b = sns.histplot(x, kde=True, ax=ax)
b.tick_params(labelsize=20)
ax.yaxis.label.set_size(20)
```

## Uniform distribution

Is a distribution where all the numbers in a certain range are equally likely to occur

```python
np.random.seed(42)
x = np.random.rand(1000)
fig, ax = plt.subplots(figsize=(20,10))
fig.suptitle('Uniform distribution', fontsize=32)
b = sns.histplot(x, kde=True, ax=ax)
b.tick_params(labelsize=20)
ax.yaxis.label.set_size(20)
```

## Log-normal distribution

Is a distribution of logarithmic values from a related normal distribution. No negative values.

```python
np.random.seed(42)
x = np.random.lognormal(3, 1, 1000)
fig, ax = plt.subplots(figsize=(20,10))
fig.suptitle('Log-normal distribution', fontsize=32)
b = sns.histplot(x, kde=True, ax=ax)
b.tick_params(labelsize=20)
ax.yaxis.label.set_size(20)
```

## Binomial distribution

Is a probability distribution where two possible outcomes are possible: Success of failure. For example, imagine that you flip a coin 10 times, what's is the probability of getting 5 heads?

```python
np.random.seed(42)
number_of_tosses = 10
probability_of_getting_head = .5 # there are 1 head and 1 tail in a fair coin, so 50% chance of getting heads
x = np.random.binomial(number_of_tosses, probability_of_getting_head, 1000)
fig, ax = plt.subplots(figsize=(20,10))
fig.suptitle('Binomial distribution', fontsize=32)
b = sns.histplot(x, kde=True, ax=ax, stat="probability", discrete=True)
b.tick_params(labelsize=20)
ax.yaxis.label.set_size(20)
```

You can see that the probability of getting 5 heads is around 0.24 or 24%.
